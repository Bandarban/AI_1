```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy as sp

%matplotlib inline
```

```python
import pandas as pd

data = pd.read_csv('C://Users/iagal/AI_1/data/painted_data.csv')
data.head()
```

```python
from numpy import NaN
nan_data = data.copy()
```

```python
nan_data_fill_simple = nan_data.fillna(0)
nan_data_fill_simple.head(20)
```

```python
nan_data_fill_simple.head()
```

```python
import sklearn.preprocessing as preprocessing

nan_data = pd.get_dummies(nan_data_fill_simple)
nan_data.head()

```

```python


mm_scalar = preprocessing.MinMaxScaler()
mm_scalar.fit(nan_data)
mm_norm_nan_data = pd.DataFrame(columns=nan_data.columns, data=mm_scalar.transform(nan_data))
mm_norm_nan_data.head()
mm_norm_nan_data.plot(figsize=(10, 10))
```
